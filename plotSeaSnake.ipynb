```python
import os
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.dates import DateFormatter
import numpy as np
from scipy import stats
import datetime
from datetime import datetime as dt
```

```python
def readData(basepath, header, startTime = False, endTime = False):
    df_SSdata = pd.DataFrame()
    l_basepath = os.listdir(basepath)
    l_basepath.sort()
    for day in l_basepath:
        pathToHour = os.path.join(basepath, day)
        #print(day)
        if os.path.isdir(pathToHour):
            l_hour = os.listdir(pathToHour)
            l_hour.sort()
            #print(l_hour)
            for hour in l_hour:
                pathToFile = os.path.join(pathToHour, hour)

                print(pathToFile)
                with open(os.path.join(pathToFile)) as currentFile:
                    df_currentData = pd.read_csv(currentFile, index_col=0,
                                            parse_dates=True, names = header,
                                            on_bad_lines='warn')
                    '''
                    df_currentData = pd.read_csv(currentFile, index_col=0,
                                            parse_dates=True, names = header,
                                            on_bad_lines='warn')
                    '''
                    #print(df_currentData["SST"].index[0])
                    #print(df_currentData["SST"].index[-1])
                df_SSdata = pd.concat([df_SSdata, df_currentData], axis = 0)
    return df_SSdata

def createDayNightData(startDateTime, EndDateTime, sunrise=360, sunset=1080,):
    print("Creating Daylight")
    currentDateTime = startDateTime
    l_dateTimes = []
    l_daylight = []
    while currentDateTime <= EndDateTime:

        currentMinuteOfDay = (currentDateTime.hour*60
                              + currentDateTime.minute
                              + currentDateTime.second/60)
        if sunrise < currentMinuteOfDay < sunset:
            daylight = 0
        else:
            daylight = 40
        l_daylight.append(daylight)
        l_dateTimes.append(currentDateTime)
        currentDateTime += datetime.timedelta(minutes=1)
    df_daylight = pd.DataFrame(l_daylight, index=l_dateTimes)
    return df_daylight, l_daylight, l_dateTimes
```
