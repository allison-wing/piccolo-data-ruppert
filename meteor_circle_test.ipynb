```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npts = 100
twopi = 2*np.pi
```

```python
# Test parameters
ship_v = (np.arange(20)+1)/20*10 # kts
kts2mps = 0.514
ship_v*=kts2mps # kts --> m/s

t_perscan = 5 # minutes per scan cycle
t_perscan*=60 # min --> s
t_total = t_perscan*4
```

```python
# track length
r = (ship_v * t_total)/twopi
circumf = r*twopi
# print('Radius [km]: ',r*1e-3)
# print('Diameter [km]: ',2*r*1e-3)
# print('Circumf [km]: ',circumf*1e-3)
```

```python
fix, ax = plt.subplots()
ax.plot(r/1e3, ship_v/kts2mps)
ax.set_xlabel('radius [km]')
ax.set_ylabel('ship velocity [kts]')
```
